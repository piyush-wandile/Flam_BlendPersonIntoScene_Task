# Running Shadow Detection and Classification on BG Image

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import io

# Step 1: Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Take the first uploaded image

# Step 2: Load and convert image
image = Image.open(io.BytesIO(uploaded[image_path])).convert('RGB')
img_rgb = np.array(image)
img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Step 3: Blur and threshold to detect shadows
blurred = cv2.GaussianBlur(gray, (7, 7), 0)
_, binary_shadow = cv2.threshold(blurred, 60, 255, cv2.THRESH_BINARY_INV)

# Step 4: Clean mask with morphology
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
cleaned = cv2.morphologyEx(binary_shadow, cv2.MORPH_OPEN, kernel)

# Step 5: Extract hard/soft shadows
edges = cv2.Canny(cleaned, 30, 100)
dilated_edges = cv2.dilate(edges, None, iterations=1)
soft_shadows = cv2.subtract(cleaned, dilated_edges)

# Step 6: Show results
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("All Shadows (Binary Mask)")
plt.imshow(cleaned, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Hard Shadows (Edges)")
plt.imshow(dilated_edges, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Soft Shadows (No Edges)")
plt.imshow(soft_shadows, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
